### Putting labels on each sample with either high or low SME (sleep maintenance efficiency) -> could divide into "better" and "worse" sleepers

```python
import pandas as pd
```

```python
df = pd.read_csv("features.csv")
```

```python
df['SME'].describe()
```

```python
df["SME"].hist(bins=20)
```

```python
# Setting 80% as threshold for "healthy sleep
sme_labels = df[['dataset', 'file', 'SME']]
sme_labels['normal_sme'] = sme_labels['SME'] >= 80
```
