```python
all_features_df = df.drop(columns=['dataset', 'file', 'UNS', 'Unnamed: 0'])
```

```python
all_features_df
```

```python
corr_matrix = all_features_df.corr().abs()
```

```python
corr_matrix
```

```python
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
upper
```

```python
# Stack and clean
correlations = (upper.stack()
                     .reset_index()
                     .rename(columns={'level_0': 'Feature 1', 'level_1': 'Feature 2', 0: 'Correlation'}))

# Sort by absolute correlation descending
correlations = correlations.iloc[correlations['Correlation'].abs().argsort()[::-1]]

print(correlations.to_string(index=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=False,          # Show correlation values in cells
    cmap="coolwarm",     # Color scheme (blue=negative, red=positive)
    vmin=-1, vmax=1,     # Fix color scale to [-1, 1]
    linewidths=0.5,      # Add gridlines between cells
)

plt.title("Feature Correlation Matrix", fontsize=14)
plt.xticks(rotation=90, ha="right")  # Rotate x labels for readability
plt.tight_layout()
plt.show()
```

```python
# remove highly correlated features
corr_threshold = 0.7 # need to select threshold - with 0.7 we keep 25 features

to_drop = [col for col in upper.columns if any(upper[col] > corr_threshold)]
selected_features = [col for col in all_features_df.columns if col not in to_drop]
print("correlation filter selected:", selected_features)
print("n =", len(selected_features))
```

```python
# then idea would be to cluster on: 1) hypno_features, 2) signal_features, 3) selected_features
```

### Hypnogram features

```python
X_hypno = hypno_feature_df.iloc[:,:-1]
y_hypno = hypno_feature_df.iloc[:,-1]
```

```python
print(X_hypno.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_hypno)
```

```python
# UMAP applied to hypnogram features to reduce dimensionality to 2

umap_2d = umap.UMAP(n_components=2, n_neighbors=3, min_dist=0.1, random_state=RANDOM_STATE)
X_umap_hypno = umap_2d.fit_transform(X_scaled)

print(X_umap_hypno.shape)
```

```python
n_clusters = 2

# agglomerative clustering

agglo = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
labels_agglo = agglo.fit_predict(X_umap_hypno)

# K-means

kmeans = KMeans(n_clusters=n_clusters, random_state = 42, n_init = 5)
labels_kmeans = kmeans.fit_predict(X_umap_hypno)

clusterings = {
    "K-means": labels_kmeans,
    "Agglomerative": labels_agglo
}
```

```python
labels_agglo
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 7))

for ax, (name, labels) in zip(axes.ravel(), clusterings.items()):
    labels = np.array(labels)

    # Plot non-noise points first
    mask_regular = labels != -1
    sc = ax.scatter(
        X_umap_hypno[mask_regular, 0],
        X_umap_hypno[mask_regular, 1],
        c=labels[mask_regular],
        alpha=0.7
    )

    # Highlight noise / outliers if present
    mask_noise = labels == -1
    if np.any(mask_noise):
        ax.scatter(
            X_umap_hypno[mask_noise, 0],
            X_umap_hypno[mask_noise, 1],
            marker="x",
            s=60,
            label="Noise / outlier"
        )
        ax.legend()

    ax.set_title(name)
    ax.set_xlabel("Dim 1")
    ax.set_ylabel("Dim 2")
    ax.grid(True)

plt.tight_layout()
plt.show()
    
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 7))

marker_map = {
    'sc': ".",
    'st': "+"}

for ax, (name, labels) in zip(axes.ravel(), clusterings.items()):
    labels = np.array(labels)

    # Plot each true label with a different marker
    for y_val in np.unique(y_hypno):

        marker = marker_map.get(y_val, "o")

        # Regular clustered points
        mask_regular = (labels != -1) & (y_hypno == y_val)

        ax.scatter(
            X_umap_hypno[mask_regular, 0],
            X_umap_hypno[mask_regular, 1],
            c=labels[mask_regular],
            marker=marker,
            alpha=0.7
        )

        # Noise points
        mask_noise = (labels == -1) & (y_hypno == y_val)

        if np.any(mask_noise):
            ax.scatter(
                X_umap_hypno[mask_noise, 0],
                X_umap_hypno[mask_noise, 1],
                marker="x",
                s=60,
                color="black"
            )
    
    # Custom legend with black symbols
    legend_elements = [
        Line2D(
            [0], [0],
            marker=marker_map.get(y_val, "o"),
            color="black",
            linestyle="None",
            markersize=8,
            label=f"{label_map[y_val]}"
        )
        for y_val in np.unique(y_hypno)
    ]

    ax.legend(handles=legend_elements)

    ax.set_title(name)
    ax.set_xlabel("Dim 1")
    ax.set_ylabel("Dim 2")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# plot based on sleep maintenance efficiency
## label data as above or below median

sme_binary = (hypno_feature_df["SME"] > hypno_feature_df["SME"].median()).astype(int)

fig, axes = plt.subplots(1, 2, figsize=(12, 7))

marker_map = {
    0: ".",
    1: "+"}

for ax, (name, labels) in zip(axes.ravel(), clusterings.items()):
    labels = np.array(labels)

    # Plot each true label with a different marker
    for y_val in np.unique(sme_binary):

        marker = marker_map.get(y_val, "o")

        # Regular clustered points
        mask_regular = (labels != -1) & (sme_binary == y_val)

        ax.scatter(
            X_umap_hypno[mask_regular, 0],
            X_umap_hypno[mask_regular, 1],
            c=labels[mask_regular],
            marker=marker,
            alpha=0.7,
            label=f"{label_map_sme[y_val]}"
        )

        # Noise points
        mask_noise = (labels == -1) & (sme_binary == y_val)

        if np.any(mask_noise):
            ax.scatter(
                X_umap_hypno[mask_noise, 0],
                X_umap_hypno[mask_noise, 1],
                marker="x",
                s=60,
                color="black"
            )

    # Custom legend with black symbols
    legend_elements = [
        Line2D(
            [0], [0],
            marker=marker_map.get(y_val, "o"),
            color="black",
            linestyle="None",
            markersize=8,
            label=f"{label_map_sme[y_val]}"
        )
        for y_val in np.unique(sme_binary)
    ]

    ax.legend(handles=legend_elements)

    ax.set_title(name)
    ax.set_xlabel("Dim 1")
    ax.set_ylabel("Dim 2")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# silhouette score

results = []

for name, labels in clusterings.items():
    labels = np.array(labels)

    # For DBSCAN, exclude noise points for silhouette if needed
    if np.any(labels == -1):
        mask = labels != -1
        X_eval = X_umap_hypno[mask]
        labels_eval = labels[mask]
    else:
        X_eval = X_umap_hypno
        labels_eval = labels

    # Silhouette is only defined if there are at least 2 clusters
    unique_clusters = set(labels_eval)
    if len(unique_clusters) >= 2 and len(labels_eval) > len(unique_clusters):
        sil = silhouette_score(X_eval, labels_eval)
    else:
        sil = np.nan

    n_clusters_found = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = np.sum(labels == -1)

    results.append({
        "Method": name,
        "Clusters found": n_clusters_found,
        "Noise points": n_noise,
        "Silhouette score": sil
    })

results_df = pd.DataFrame(results).sort_values(by="Silhouette score", ascending=False)
print(results_df)
```
