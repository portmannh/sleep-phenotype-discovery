```python
import mne
import glob
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
import yasa
```

```python
raw = mne.io.read_raw_fif('/home/alexia/Downloads/sleep-edf-database-expanded-1.0.0/sleep-cassette/preprocessed/SC4001E0-raw.fif', preload=True, verbose='ERROR')
```

```python
eeg = raw.copy().pick_channels(['EEG Fpz-Cz', 'EEG Pz-Oz'])
eeg.load_data()
```

```python
eeg.info['sfreq']
```

```python
# build hypnogram
hypno = eeg.annotations.description
hyp = yasa.Hypnogram.from_integers(hypno, mapping={0: 'W', 1: 'N1', 2: 'N2', 3: 'N3', 4: 'R', -1: 'Uns'}, n_stages=5, freq='30s')
```

```python
# detect slow waves
sw = yasa.sw_detect(eeg, hypno=hyp, include=(2,3))
```

```python
df = sw.summary(grp_chan=True).reset_index()
```

```python

df_wide = df.pivot_table(index=None, columns='Channel', values='Count', aggfunc='first')
df_wide = df_wide.reset_index(drop=True)
df_wide.columns = [f'{col}_SW_density' for col in df_wide.columns]
```
